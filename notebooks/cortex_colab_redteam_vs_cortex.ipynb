# Cortex Red Team vs Cortex Colab Lab

Notebook Google Colab complet pour simuler une Red Team AI face a Cortex.

Ce notebook ne prend jamais la decision finale de securite. Il produit seulement des signaux exploitables pour Cortex.

## Ce que le notebook couvre

- environnement dynamique users / devices / services / resources / sessions
- graphe entreprise + GNN leger pour risque de noeud
- Red Team autonome avec actions offensives
- agents Cortex cooperatifs: Sentinel, Trust Engine, Threat Hunter, Graph Analyst, Anomaly Detector
- anticipation de la prochaine action via sequence model LSTM
- double reinforcement learning: attaquant et defense
- signaux pour zero-day, low-and-slow, multi-step et prepositionnement
- visualisation, evaluation et export des artefacts
